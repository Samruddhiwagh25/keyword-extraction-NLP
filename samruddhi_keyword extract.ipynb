### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Reading File into dataframe

```python
df=pd.read_json('../travee/dataset.json')
```

```python
df.head()
```

```python
#Taking out feedback reviews from the feedback column and appending into new list
data=[]
for i in range(0,df.shape[0]):
    data.append(df['feedback'].iloc[i])
```

```python
#concating all the list into single list
concatenated_list = [item for sublist in data for item in sublist]
```

```python
df_fb=pd.DataFrame(data=concatenated_list)
```

```python
df_fb.head()
```

```python
df_new=df[['pid','name']]  #creating new dataframe of only pid and hotel names 
```

```python
df_fb_new=df_fb[['loc_id','free_text']]  #creating new dataframe of only loc_d and free_text 
```

```python
df_fb_new
```

```python
df_new
```

### Left Joining both the data Frames

```python
result=pd.merge(df_new,df_fb_new,how='left',left_on='pid',right_on='loc_id')
```

```python
result=result[['name','free_text']]  #taking out only hotel names and reviews from the joined table
```

```python
result
```

```python
#creating new data frame by concatinating all the reviews of the same hotels
hotel_feedback = result.groupby('name')['free_text'].apply(lambda x: ' '.join(x.dropna())).reset_index()
```

```python
hotel_feedback  #new concatinated reviews dataframe
```

### importing keybert library for keyword extraction

```python
from keybert import KeyBERT  
```

```python
#creating model and appending keywords into list
a=[]
kw_model=KeyBERT()
for i in range(0,12):
    keywords=kw_model.extract_keywords(hotel_feedback['free_text'].iloc[i],keyphrase_ngram_range=(1,1),use_maxsum=True,top_n=10)
    a.append(keywords)
```

```python
(a)
```

```python
b=list(hotel_feedback['name'])
```

```python
#creating dictionary for key value pairs of hotel names and their keywords
my_dict = dict(zip(b,a))
```

```python
my_dict
```

```python
#taking out only the keywords from the inside value tuples 
new_dict = {hotel: [item[0] for item in items if item] for hotel, items in my_dict.items()}
```

```python
new_dict
```

```python
#creating final dataFrame of hotels and keywords
submit = pd.DataFrame({'Hotels':hotel_feedback['name'],'Keywords':new_dict.values()})
```

```python
submit
```

### Saving the keywords into csv file

```python
submit.to_csv('Samruddhi_keywords_extracted.csv',index=False)
```
